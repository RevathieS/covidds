```python
# top 5 countries which have more number of covid cases (hint use df.nlargest)
top_countries_covid_cases = df.nlargest(5, 'Total Cases')
top_countries_covid_cases
```

```python
#plot bar graph for top 5 countries
plt.figure(figsize=(10, 6))
plt.bar(top_countries_covid_cases['Country'], top_countries_covid_cases['Total Cases'], color='skyblue')
plt.title('Top 5 Countries with the Most COVID Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# least 5 countries which have least number of covid cases
least_countries_covid_cases = df.nsmallest(5,'Total Cases')
least_countries_covid_cases
```

```python
#draw bar graph of least 5 countries
plt.figure(figsize=(10, 6))
plt.bar(least_countries_covid_cases['Country'], least_countries_covid_cases['Total Cases'], color='lightcoral')
plt.title('Least 5 Countries with the Least COVID Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)#
plt.show()
```

```python
#top 5 countries which have more number of covid Deaths
more_death = df.nlargest(5,"Total Deaths") 
more_death
```

```python
#plot bar graph for top 5 countries which have more number of covid Deaths
plt.figure(figsize=(10, 6))
plt.bar(more_death['Country'], more_death['Total Cases'], color='lightcoral')
plt.title('Top 5 Countries with the Most COVID Cases Death')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
#5 countries which have least number of covid Death
less_death = df.nsmallest(5,"Total Deaths") 
less_death
```

```python
#Death percentage
#top 5 countries which have more number of covid  Death percentage 
death_percentage= df.nlargest(5,"Death percentage")
death_percentage
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
plt.figure(figsize=(10, 6))
plt.bar(death_percentage['Country'], death_percentage['Total Cases'], color='lightcoral')
plt.title('Top 5 Countries with the most COVID Cases Death Percentage')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
#continent wise cases (hint use group by)
continent_wise_cases = df.groupby('Continent')['Total Cases'].sum().reset_index()
continent_wise_cases
```

```python
#draw bar plot for continent wise case
plt.figure(figsize=(10, 6))
plt.bar(continent_wise_cases['Continent'], continent_wise_cases['Total Cases'], color='green')
plt.title('Continent-wise COVID Cases')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.show()
```

```python
#find relationship between Population and total cases.Comment  on outliers
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='Total Cases', data=df)
plt.title('Relationship between Population and Total Cases')
plt.xlabel('Population')
plt.ylabel('Total Cases')
plt.show()

# Calculate correlation coefficient
correlation_coefficient = df['Population'].corr(df['Total Cases'])
print(f'Correlation Coefficient: {correlation_coefficient}')


plt.figure(figsize=(10, 6))
sns.boxplot(x='Total Cases', data=df)
plt.title('Box Plot of Total Cases')
plt.show()
```
